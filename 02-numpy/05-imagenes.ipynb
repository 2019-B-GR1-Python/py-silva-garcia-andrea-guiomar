```python
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
mapache   #RGB
```

```python
mapache.shape
```

```python
print(type(mapache)) #arreglo de numpy  
```

```python
#dibujar al mapache
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
```

```python
plt.imshow(mapache, cmap=plt.cm.gray, vmin=30, vmax=200)
```

```python
plt.imshow(mapache[200:220, 200:220], cmap=plt.cm.gray)
```

```python
plt.imshow(mapache[::-1])
```

```python
plt.imshow(mapache[200:220, 200:220], cmap=plt.cm.gray, interpolation='nearest')
```

```python
mx, my, mz = mapache.shape

flip_ud_mapache = np.flipud(mapache)
plt.imshow(flip_ud_mapache, cmap=plt.cm.gray)
```

```python
rotate_mapache = ndimage.rotate(mapache, 45)
plt.imshow(rotate_mapache, cmap=plt.cm.gray)
```

```python
rotate_mapache_noreshape = ndimage.rotate(mapache, 45, reshape=False)
plt.imshow(rotate_mapache_noreshape, cmap=plt.cm.gray)
```

```python
local_mean = ndimage.uniform_filter(mapache, size=11)
plt.imshow(local_mean, cmap=plt.cm.gray)
```

```python
very_blurred = ndimage.gaussian_filter(mapache, sigma=5)
plt.imshow(very_blurred, cmap=plt.cm.gray)
```

```python
blurred_mapache = ndimage.gaussian_filter(mapache, sigma=3)
plt.imshow(blurred_mapache, cmap=plt.cm.gray)
```
